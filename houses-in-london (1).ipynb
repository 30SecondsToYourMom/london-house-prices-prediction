```python
import numpy as np 
import pandas as pd 
```

```python
data = pd.read_csv('london_houses.csv')
```

```python
data
```

```python
data.info()
```

```python
data["Address"] = data['Address'].apply(lambda x: ' '.join(x.split(' ')[1:]))
```

```python
data['Address']
```

```python
data = pd.get_dummies(data)
```

```python
data
```

```python
from sklearn.linear_model import ElasticNetCV
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import SVR
from sklearn.ensemble import AdaBoostRegressor, GradientBoostingRegressor, RandomForestRegressor
```

```python
X = data.drop('Price (£)', axis=1)
y = data['Price (£)']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import Pipeline, make_pipeline
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
linear_model = ElasticNetCV()
```

```python
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
linear_model.fit(X_train,y_train)
```

```python
pred = linear_model.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
print(mean_absolute_error(y_test, pred))
```

```python
print(np.sqrt(mean_squared_error(y_test,pred)))
```

```python
def run_model(model, X_train, X_test, y_train, y_test):
    model.fit(X_train,y_train)
    pred = model.predict(X_test)
    print(mean_absolute_error(y_test, pred))
    print(np.sqrt(mean_squared_error(y_test,pred)))
```

```python
knn = run_model(KNeighborsRegressor(), X_train, X_test, y_train, y_test)
```

```python
svr = run_model(SVR(),X_train, X_test, y_train, y_test)
```

```python
random_forest = run_model(RandomForestRegressor(), X_train, X_test, y_train, y_test)
```

```python
ada = run_model(AdaBoostRegressor(), X_train, X_test, y_train, y_test)
```

```python
gradient = run_model(GradientBoostingRegressor(), X_train, X_test, y_train, y_test)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
grid_param = {'n_estimators':[500,1000,1500],
             'max_depth':[2,3,4]}
```

```python
grid = GridSearchCV(GradientBoostingRegressor(), grid_param, cv=5, scoring='neg_mean_absolute_error')
```

```python
grid.fit(X_train,y_train)
```

```python
print("best_params:", grid.best_params_)
print("best_score:", grid.best_score_)
```

```python
results = grid.cv_results_
for mean_score, params in zip(results['mean_test_score'], results['params']):
    print(f"MAE: {-mean_score:.4f}, Params: {params}")
```
